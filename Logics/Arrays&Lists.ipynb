```python
'''
Find the Unique Element occured only once
Input: 
Testcase
N
Arr

1
7
2 3 1 6 3 6 2

Output:
Unique Element

1
'''
```

```python
t = int(input())

for i in range(t):
    
    n = int(input())
    arr = input().split()
    res_dict = {}
    for c in arr:
        
        if c in res_dict:    
            res_dict[c] += 1
        else:
            res_dict[c] = 1
    
    for key, value  in res_dict.items():
        if value == 1:
            print(key)
            break
 
```

```python
'''
You have been given an integer array/list(ARR) and a number X. 
Find and return the total number of pairs in the array/list which sum to X.
Input:
1
9
1 3 6 2 5 4 3 2 4
7
Output:
7
'''
t = int(input())
pair_count = 0
for _ in range(t):
    N = int(input())
    arr = []
    if N > 0:
        arr = list(map(int, input().split()))
    pair_sum = int(input())
    for i in range(N):
        for j in range(i + 1, N):
            if (arr[i] + arr[j]) == pair_sum:
                pair_count += 1
    print(pair_count)
    

```

```python
'''
Given an array of N elements switch(swap) the element with the adjacent element and print the output.
Sample Testcase :
INPUT
5
3 2 1 2 3
OUTPUT
2 3 2 1 3
'''
#A Simple Hello World

#Getting input via STDIN
userInput = input()

arr = userInput.split()

res_arr = []

arr_len = len(arr)

for i in range(0, arr_len, 2):
    if i + 1 != arr_len:
        res_arr.append(arr[i+1])
        res_arr.append(arr[i])
    else:
        res_arr.append(arr[i])
res_arr
```
